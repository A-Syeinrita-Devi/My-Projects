Several techniques can be used to handle missing values effectively instead of just removing rows with missing values:

- Imputation: Imputation entails estimating missing values using the information that is currently available. Mean, median, mode, and predictive imputation using regression models or machine learning algorithms are examples of common imputation techniques. We can fill in the missing values using imputation while keeping the current observations.


- Substitution: If appropriate for the variable and the context, missing values may occasionally be replaced with a particular value or category, such as "Unknown" or "Not Applicable."


- Multiple imputation: To account for uncertainty and provide a range of potential values, multiple imputation generates multiple plausible values for each missing entry. When uncertainty in the imputed values must be taken into account in subsequent analyses, this method is helpful.


- Time-series interpolation: For time-series data, forward-fill (ffill) or backward-fill (bfill) interpolation techniques can be used to fill in missing values using the series' prior or subsequent values.


- Formulas: When there are logical relationships or patterns in the dataset, formulas or calculations can be used to estimate missing values. The formulas employed must, however, be suitable for the particular context and variables being examined.


We can ensure the integrity and quality of our data analysis or modelling and draw more trustworthy and accurate conclusions from the data by handling missing values properly.


Interpolation may be a good option to handle missing values because the given data for the project are time series data. A technique called interpolation calculates missing data point values based on the available observations. It functions by substituting values that are derived from nearby data points or the overall trend of the time series for the missing values.  It is a method that is frequently used in time series analysis and can support the accuracy and dependability of further data analyses and modelling.


Using interpolation to handle missing values in time series data has several benefits:

1. Preservation of temporal relationships: The temporal ordering of the data points is taken into account during interpolation. It keeps the time series' natural relationships and patterns by estimating missing values based on the nearby observations. This is especially significant in time series analysis, where understanding trends, seasonality, and other time-dependent patterns depends greatly on the order and timing of data points.


2. Data continuity retention: Interpolation aids in preserving the time series' continuity. It ensures that the time series is uninterrupted and complete by adding any missing values. This is helpful when examining the general trends and traits of the time series over a specific time frame.


3. Consistency with the overall trend: Interpolation techniques take into account the time series' overall trend and pattern. The estimated values are in agreement with the overall pattern of the time series because they take into account nearby data points. This makes it possible to represent the underlying trend more accurately, which is useful for forecasting, trend analysis, and other time-dependent analyses.


4. Statistical effectiveness: To estimate missing values, interpolation makes use of the information at hand. In comparison to simply removing missing data points, it makes the most of the information that is already available and can produce estimates that are more statistically accurate.

## DATA AGGREGATION & GROUP OPERATIONS

```python
# Assuming df2 is the original dataframe with columns 'Year', and other columns

# Convert 'Year' column to datetime type
df1['Year'] = pd.to_datetime(df1['Year'], format='%Y')

# Filter data starting from 1970
df1 = df1[df1['Year'].dt.year >= 1960]

# Group the data by decade and calculate the mean for each group
df_fertility_rate_decades = df1.groupby(pd.Grouper(key='Year', freq='10Y')).mean().reset_index()

# Extract the year from the 'Year' column
df_fertility_rate_decades['Year'] = df_fertility_rate_decades['Year'].dt.year

# Rename the column
df_fertility_rate_decades.rename(columns={'Year': 'Year_Group_10years'}, inplace=True)

# Convert all columns to integer type
df_fertility_rate_decades = df_fertility_rate_decades.astype(int)

# Print the grouped dataframe
df_fertility_rate_decades
```

On the dataframe df1, operations for grouping and aggregating data were carried out. The data were divided into 10-year intervals with the intention of calculating the mean values for each group. The 'Year' column was removed from the datetime format using.dt.year and renamed to 'Year_Group_10years' using rename() in order to make the dataframe more readable and informative. The column now clearly states that it represents 10-year intervals thanks to the renaming. The astype() function was utilised to convert all columns in the dataframe to integers in order to align with the original dataset and guarantee the data type of each column in the dataframe.

The data has been grouped and aggregated, and the resulting dataframe, df_fertility_rate_decades, contains the mean values for each 10-year period. For examining fertility rate trends and patterns over time, this data can be helpful. In order to provide a balance between granularity and reducing noise from year-to-year fluctuations, it was decided to divide the data into 10-year intervals. This allowed for a clearer understanding of longer-term trends.

```python
from statsmodels.graphics.tsaplots import plot_acf

# Plotting the ACF
plt.figure(figsize=(12, 6))
plot_acf(df1['Total Fertility Rate'], lags=20)
plt.xlabel('Lag')
plt.ylabel('Autocorrelation')
plt.title('Autocorrelation Function (ACF)')
plt.grid(True)
plt.show()
```

We are able to learn more about the stationarity of the data by looking at the autocorrelation function (ACF) plot. When the ACF shows significant autocorrelation values over a range of lags, non-stationarity in the data is likely present.

If we notice a slow decline in the autocorrelation values in the provided ACF plot, it suggests that the data is not stationary. The presence of trends or seasonality is indicated by a slow decay in autocorrelation, which also suggests a strong dependence on historical values.

According to the ACF plot, differencing is required because the data are not stationary. By removing underlying trends or seasonality, differencing can make data stationary. By using differencing, we can change time series into new first differences series, which should have less autocorrelation and a mean and variance that are more constant.


```python
# Compute first-order differences
df1_diff = df1['Total Fertility Rate'].diff().dropna()

# Create subplots
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Plot the first-order differences
axes[0].plot(df1_diff)
axes[0].set_xlabel('Year')
axes[0].set_ylabel('First-Order Differences')
axes[0].set_title('First-Order Differences of Total Fertility Rate')
axes[0].grid(True)

# Plot the ACF of the first-order differences
plot_acf(df1_diff, lags=20, ax=axes[1])
axes[1].set_xlabel('Lag')
axes[1].set_ylabel('Autocorrelation')
axes[1].set_title('Autocorrelation Function (ACF) of First-Order Differences')
axes[1].grid(True)

# Adjust the layout and spacing between subplots
plt.tight_layout()

# Display the plots
plt.show()
```

First-order differences are plotted in the first subplot. We eliminate the trend and any seasonality in the data by taking the difference between consecutive values. As a result, there should be fewer obvious trends or patterns in the plot that shows the changes between adjacent observations.

The first-order differences' ACF is shown in the second subplot. We can use it to evaluate the differenced series' autocorrelation. Since the ACF values rapidly decrease and fall within the significance bounds (horizontal dashed lines), differencing has likely removed the majority of the autocorrelation, making the data more stationary.


This plot's objective is to demonstrate whether the data have been successfully converted into a stationary series through the differencing operation. The differenced series' ACF plot exhibits a significantly lower autocorrelation than the original series, indicating that differencing has improved the data's suitability for modelling with ARIMA.


```python
# Compute second-order differences
df1_diff2 = df1_diff.diff().dropna()

# Create subplots
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Plot the second-order differences
axes[0].plot(df1_diff2)
axes[0].set_xlabel('Year')
axes[0].set_ylabel('Second-Order Differences')
axes[0].set_title('Second-Order Differences of Total Fertility Rate')
axes[0].grid(True)

# Plot the ACF of the second-order differences
plot_acf(df1_diff2, lags=20, ax=axes[1])
axes[1].set_xlabel('Lag')
axes[1].set_ylabel('Autocorrelation')
axes[1].set_title('Autocorrelation Function (ACF) of Second-Order Differences')
axes[1].grid(True)

# Adjust the layout and spacing between subplots
plt.tight_layout()

# Display the plots
plt.show()
```

The data is subjected to a second differentiation, and by comparing the two plots, we can judge whether the second differentiation was successful in achieving stationarity and ascertain whether the transformed data is suitable for ARIMA modelling.


```python
from statsmodels.tsa.stattools import adfuller

# Perform ADF test on the original data
adf_before = adfuller(df1['Total Fertility Rate'])
print('ADF test statistics before differencing:', adf_before[0])
print('p-value before differencing:', adf_before[1])
print('Critical values before differencing:')
for key, value in adf_before[4].items():
    print(f'{key}: {value}')

print()    
    
# Perform ADF test on the differenced data
adf_after = adfuller(df1_diff)
print('ADF test statistics after differencing:', adf_after[0])
print('p-value after differencing:', adf_after[1])
print('Critical values after differencing:')
for key, value in adf_after[4].items():
    print(f'{key}: {value}')

print()
    
    # Perform ADF test on the differenced 2 data
adf_after = adfuller(df1_diff2)
print('ADF test statistics after 2nd differencing:', adf_after[0])
print('p-value after differencing:', adf_after[1])
print('Critical values after differencing:')
for key, value in adf_after[4].items():
    print(f'{key}: {value}')
```

The stationarity of a time series can be ascertained statistically using the Augmented Dickey-Fuller,ADF test. For each case, the ADF test statistics, p-values, and critical values at various levels of significance (1%, 5%, and 10%) are printed. The ADF test statistics gauge how stationary the data is, and the p-value denotes the likelihood that the test statistic will be obtained if the null hypothesis of non-stationarity is true.

The p-value must be below a predetermined threshold of significance (for example, 0.05) in order for a time series to be deemed stationary, indicating strong evidence against the null hypothesis of non-stationarity. In this instance, we can see that the p-value drops dramatically (nearly to 0) after the first-order differencing, indicating that the differenced data is very likely to be stationary. At all levels of significance, the ADF test statistics are also appreciably below the critical values.

The p-value also remains extremely low after the second-order differencing, suggesting that the second-order differenced data is also most likely to be stationary. In comparison to the critical values, the ADF test statistics are significantly lower.


These findings demonstrate that the data have become stationary as a result of the differencing operations, which is a prerequisite for using ARIMA modelling. ARIMA is a useful model for forecasting because the stationarity of the data enables us to use autoregressive (AR) and moving average (MA) components to capture the temporal dependencies.

Autocorrelation Function (ACF) and Partial Autocorrelation Function (PACF) for the first and second differencing of the data are plotted. 

ACF- The correlation between the current observation and its lag values is displayed by the ACF plot. It aids in determining the Moving Average (MA) component's position in the ARIMA model. Significant spikes outside of the confidence interval on the ACF plot signify the existence of significant correlations at those lags.

PACF - The PACF plot, on the other hand, accounts for correlations at shorter lags and displays the correlation between the current observation and its lagged values. It aids in determining the ARIMA model's Autoregressive (AR) component's position. In the PACF plot, significant spikes outside the confidence interval denote the presence of significant correlations at those lags, just like in the ACF plot.


We have three main parameters in an ARIMA model: p, d, and q.

The parameter p (AR order) indicates the position of the Autoregressive (AR) component in the model. It establishes how many lag observations should be used in the model. The Partial Autocorrelation Function (PACF) plot, which accounts for shorter lags and displays the correlation between the current observation and its lagged values, can be used to calculate the value of p. The significant spikes in the PACF plot show what p might be worth.


The parameter d (Differencing order) indicates the order of differencing necessary to make the time series stationary. It establishes how many times the data must be subjected to the differencing operation. We can examine the time series plot, the autocorrelation function (ACF) plot, and run statistical tests like the Augmented Dickey-Fuller (ADF) test to look for stationarity in order to calculate the value of d.

The Moving Average (MA) component of the model is represented by the parameter q (MA order), which denotes its order. It chooses how many lags in the forecast errors to include in the model. The ACF plot, which depicts the correlation between the current observation and its laggard forecast errors, can be used to calculate the value of q. Significant spikes in the ACF plot suggest that q may have some value.


In an ARIMA model, we use the analysis of stationarity and differencing to determine the differencing order (d), and the PACF plot to determine the AR order (p), the ACF plot to determine the MA order (q), and so on. In order for the ARIMA model to successfully capture the underlying patterns and correlations in the time series data, it is helpful to choose appropriate parameter values with the aid of these plots and analyses.


We can see that the first lag is the most important one in both plots. Thus, based on ACF, we take q to be 1. Since we differentiated twice, we can say that differencing, d is either 1 or 2, and according to the PACF, p can be 1, 2, or 3. We can interpret the results using a few of the measures as shown below to determine which parameter combination is the best.


```python
from statsmodels.tsa.arima.model import ARIMA
import warnings

# Ignore warnings
warnings.filterwarnings("ignore")

# Define parameter combinations
orders = [(1, 1, 1), (1, 2, 1), (2, 2, 1),(3,2,1)]

# Fit and evaluate models for each parameter combination
for order in orders:
    model = ARIMA(df1['Total Fertility Rate'], order=order)
    model_fit = model.fit()
    
    print(f"ARIMA{order} Model Summary:")
    print(f"AIC: {model_fit.aic}")
    print(f"BIC: {model_fit.bic}")
    print(f"Log Likelihood: {model_fit.llf}")
    print(f"HQIC: {model_fit.hqic}")
    print()
```

AIC (Akaike Information Criterion), BIC (Bayesian Information Criterion), log-likelihood, and HQIC (Hannan-Quinn Information Criterion) are a few of the criteria we can use to select the best-fit model in ARIMA. These criteria offer indicators of the model's complexity and goodness of fit.

A better fit is indicated by lower values of AIC, BIC, and HQIC as well as higher log-likelihood. Finding the right balance between model complexity and performance is crucial. 

The ARIMA(1, 2, 1) model has the lowest AIC, BIC, and HQIC values when compared to the other models, indicating a better fit when considering the information that has been provided. Additionally, compared to the other models, it has a higher log-likelihood. The ARIMA(1, 2, 1) model is therefore the option that fits the data in this situation the best.


```python
from statsmodels.tsa.arima.model import ARIMA

# Fit ARIMA(1,1,1) model
model2 = ARIMA(df1['Total Fertility Rate'], order=(1, 2, 1))
model2_fit = model2.fit()

# Print model summary
print(model2_fit.summary())
```
