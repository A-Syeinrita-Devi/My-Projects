#### Question 2(g)
Consider following rules in classify the attendance:

    Rate > 90% : Good
    60% < Rate < 90% : Moderate
    Rate < 60% : Bad

Based on the grouping rule above, get the status for each of the staff and save the status in a new column `status` in the Data Frame


```python
# WRITE YOUR CODE HERE

def status(rate):
    
    if rate >90:
        status= 'Good'
    elif rate <60:
        status= 'Bad'
    else:
        status= 'Moderate'
    return status
    

emp_attendance['status'] = emp_attendance['rate'].apply(status)
emp_attendance
```

#### Question 2(h)
Construct a function that returns the average attendance of all staff over the 14 days according to the department

```python
# WRITE YOUR CODE HERE

# WRITE YOUR CODE HERE

def department_average_attendance(emp_attendance):
    return emp_attendance.groupby(['Department'],as_index=False).total.mean()

print(department_average_attendance(emp_attendance))
```
