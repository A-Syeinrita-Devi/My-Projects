#### Question 3(i)
Create a function named eliminate_ref. This function will receive an array and perform these tasks:

#### Question 3(i(i))
Find whether there is any reference attached to any string in the array given. Commonly, any references can be detected by a square bracket containing a number like [5] as shown in figure below. 

```python
# WRITE YOUR CODE HERE
import re

pattern = re.compile(r"\[(\d+)\]")

def eliminate_ref(arr):
    for r in arr:
        if len(pattern.findall(r)) >= 1:
            replcewith = r[0:(len(r)-(len(pattern.findall(r)) + 2))]
            print("Reference found: "+ r)
            arr.replace({r: replcewith}, inplace=True)

print("Category")
eliminate_ref(tallest["Category"])

print()

print("Structure")
eliminate_ref(tallest["Structure"])
```

#### Question 3 (i(ii))
If there is any reference attached, replace the word with the original word but without the reference. Set inplace=True when performing the replace method.
Test your function using these codes 


The output should be as shown in figure below. 


```python
# WRITE YOUR CODE HERE

import re

pattern = re.compile(r"\[(\d+)\]")

def eliminate_ref(column):
    for i in column.index:
        value = column.loc[i]
        matches = pattern.findall(value)
        if len(matches) >= 1:
            replace_with = re.sub(pattern, '', value)
            print("Reference found: " + value)
            column.replace({value: replace_with}, inplace=True)

eliminate_ref(tallest["Category"])
eliminate_ref(tallest["Structure"])
tallest
```
